```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2
```

# **Fetching entire Dataset (Train + Test)**

```python
path11 = '/content/drive/MyDrive/BrainDead /brain_dead_emotion_detection/Train/angry'
path21 = '/content/drive/MyDrive/BrainDead /brain_dead_emotion_detection/Train/happy'
path31 = '/content/drive/MyDrive/BrainDead /brain_dead_emotion_detection/Train/neutral'
path41 = '/content/drive/MyDrive/BrainDead /brain_dead_emotion_detection/Train/sad'
path12 = '/content/drive/MyDrive/BrainDead /brain_dead_emotion_detection/Test/angry'
path22 = '/content/drive/MyDrive/BrainDead /brain_dead_emotion_detection/Test/happy'
path32 = '/content/drive/MyDrive/BrainDead /brain_dead_emotion_detection/Test/neutral'
path42 = '/content/drive/MyDrive/BrainDead /brain_dead_emotion_detection/Test/sad'
angry1 = os.listdir(path11)
happy1 = os.listdir(path21)
neutral1 = os.listdir(path31)
sad1 = os.listdir(path41)
angry2 = os.listdir(path12)
happy2 = os.listdir(path22)
neutral2 = os.listdir(path32)
sad2 = os.listdir(path42)
print(len(angry1))
print(len(happy1))
print(len(neutral1))
print(len(sad1))
print(len(angry2))
print(len(happy2))
print(len(neutral2))
print(len(sad2))
```

```python
for img in happy1:
  X.append(cv2.imread(os.path.join(path21,img)))
  Y.append(2)
```

```python
for img in neutral1:
  X.append(cv2.imread(os.path.join(path31,img)))
  Y.append(3)
```

```python
for img in happy2:
  X.append(cv2.imread(os.path.join(path22,img)))
  Y.append(2)
```

```python
for img in neutral2:
  X.append(cv2.imread(os.path.join(path32,img)))
  Y.append(3)
```

# **Fetching the Trained Model**

```python
import joblib
RF = joblib.load('/content/drive/MyDrive/BrainDead/RF.joblib')
```

# **Getting Predictions on the given dataset**
